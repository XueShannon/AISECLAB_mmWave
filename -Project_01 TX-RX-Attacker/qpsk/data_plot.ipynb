```python
import numpy
import matplotlib.pyplot as plt
```

```python
f = numpy.fromfile(open("data/attacker.data"), dtype=numpy.complex64)
print(f)
n=100
x=f[1024*n:1024*(n+1)-1]
print(x)
plt.plot(numpy.real(x),'s-',color='r')
# plt.plot(numpy.imag(x),'o-',color='g')
plt.show()

#use numpy compute fft
m = numpy.fft.fft(x)
plt.plot(abs(m))
plt.show()
```
